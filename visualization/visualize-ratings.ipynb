# Visualizing the evolution of ratings

This short Jupyter notebook demonstrates how to use the provided `viz` library to plot the evolution of ratings over time, given the CSV output of the `ChessAnalysis` program.

```python
import viz
%matplotlib inline
```

```python
data = viz.read_csv("output.csv")
```

Find the players who have a rating for at least five years.

```python
for name, vals in data.items():
    if len(vals) >= 5:
        print(name)
```

Plot the evolution of ratings for a small sample of the players.

```python
names = (
    "Anderssen; Adolf",
    "Loewenthal; Johann Jacob",
    "Mayet; Carl",
)
fig, ax = viz.plot_skills(data, names)
```
